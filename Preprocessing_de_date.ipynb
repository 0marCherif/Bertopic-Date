```
import re
import pandas as pd
from bertopic import BERTopic

# Prepare data
trump = pd.read_csv('https://drive.google.com/uc?export=download&id=1xRKHaP-QwACMydlDnyFPEaFdtskJuBa6')
trump.text = trump.apply(lambda row: re.sub(r"http\S+", "", row.text).lower(), 1)
trump.text = trump.apply(lambda row: " ".join(filter(lambda x:x[0]!="@", row.text.split())), 1)
trump.text = trump.apply(lambda row: " ".join(re.sub("[^a-zA-Z]+", " ", row.text).split()), 1)
trump = trump.loc[(trump.isRetweet == "f") & (trump.text != ""), :]
timestamps = trump.date.to_list()
tweets = trump.text.to_list()

# Create topics over time
```

```
print(type(timestamps))
print(type(timestamps[0]))
```

```
# prompt: only keep the month and year for trump['date']

trump['date'] = pd.to_datetime(trump['date']).dt.to_period('M')
timestamps = trump.date.to_list()
```

```
timestamps2 = [str(ts) for ts in timestamps]
print(type(timestamps2))
print(type(timestamps2[0]))
```

```
model = BERTopic(verbose=True,min_topic_size=100)
topics, probs = model.fit_transform(tweets)
topics_over_time = model.topics_over_time(tweets, timestamps2)
```

```
model.visualize_topics_over_time(topics_over_time,normalize_frequency=True,topics=[14,17,28,35])
```
